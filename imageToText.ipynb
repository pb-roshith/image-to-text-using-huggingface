```python
from transformers import pipeline
```

```python
def imgToTxt(url):
    model = pipeline("image-to-text", model="Salesforce/blip-image-captioning-large")
    text = model(url)
    print(text)
```

```python
imgToTxt('hacker.jpeg')
```

```python
imgToTxt('ai.jpg')
```
